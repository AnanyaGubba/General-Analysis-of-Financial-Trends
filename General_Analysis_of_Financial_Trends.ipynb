**AAPL - CANDLESTICK CHART**

25)  

*   Each candlestick shows the opening,closing, highest, and lowest price for a given day

*   Green candles indicate a close higher than the open (bullish).


*   Shorter wicks indicate less volatility, while longer wicks indicate more.
*   The price range moves from around 145 at the start of January 2022 to above 180 by March 2022, indicating a strong price increase over this period.


*   In summary, this chart shows a positive performance for AAPL stock over the specified time, with an overall upward trend and relatively stable growth.




```python
import plotly.graph_objects as go

# Sample prices data with OHLC values
data = {
    'date': pd.date_range(start='2022-01-01', periods=100, freq='D'),
    'symbol': ['AAPL'] * 100,
    'open': [150 + i * 0.3 for i in range(100)],
    'high': [155 + i * 0.3 for i in range(100)],
    'low': [145 + i * 0.3 for i in range(100)],
    'close': [150 + i * 0.5 for i in range(100)]
}

prices_df = pd.DataFrame(data)

symbol = 'AAPL'
symbol_data = prices_df[prices_df['symbol'] == symbol]

fig = go.Figure(data=[go.Candlestick(
    x=symbol_data['date'],
    open=symbol_data['open'],
    high=symbol_data['high'],
    low=symbol_data['low'],
    close=symbol_data['close']
)])

fig.update_layout(
    title=f"Candlestick Chart for {symbol}",
    xaxis_title="Date",
    yaxis_title="Price",
    width=800,
    height=600
)
fig.show()
```

**Total Revenue vs net Income - WATERFALL CHART**

 26) Green Bars: Positive values or gains.

Red Bars: Negative values or costs that reduce the overall amount.

Blue Bar: The final value, representing net income.

* The chart shows how the total revenue is reduced by various costs to arrive at the net income. Each step represents a different expense or income factor.

* The Cost of Revenue has a significant impact, reducing the initial revenue by a noticeable amount, which is visible by the large red bar.



```python
import pandas as pd
import plotly.graph_objects as go

# Sample fundamentals data
data = {
    'Metric': ['Total Revenue', 'Cost of Revenue', 'Gross Profit', 'Operating Income', 'Net Income'],
    'Value': [100000, -50000, 50000, 20000, 15000]
}

fundamentals_df = pd.DataFrame(data)

fig = go.Figure(go.Waterfall(
    name="Financial Breakdown",
    orientation="v",
    measure=["relative", "relative", "relative", "relative", "total"],
    x=fundamentals_df["Metric"],
    textposition="outside",
    y=fundamentals_df["Value"],
    connector=dict(line=dict(color="gray", width=1))
))

fig.update_layout(
    title="Waterfall Chart: Total Revenue to Net Income",
    showlegend=True,
    width=800, height=500
)

fig.show()
```
